```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from matplotlib.colors import LogNorm
import fun.my_functions as mf
```

```python
data = np.load("../data/jica_datarequest_nr4.npz")
data.files
```

```python
spec = data['n1']
time = data['time']
e_bins = np.arange(255)
data_c = mf.red_j_spec_err(spec, [35, 63], [6, 12], savename='req4')
```

```python
fig, ax = plt.subplots(figsize=(10, 4))

im = ax.imshow(
    data_c,
    aspect='auto',
    origin='lower',
    extent=[time[0], time[-1], e_bins[0], e_bins[-1]],
    norm=LogNorm(),
    cmap='viridis'
)
fig.colorbar(im, ax=ax, label="raw integer current")
ax.set_xlabel("Time")
ax.set_ylabel("Energy Bin nr")
plt.savefig('figs/energy_time_spectrum_req4.png')
plt.show()
```
